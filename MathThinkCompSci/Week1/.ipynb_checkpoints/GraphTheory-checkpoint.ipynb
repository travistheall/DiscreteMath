```python
# hotels = pd.DataFrame([10, 15, 20], index=['a','b','c'], columns=['v'])
# can be rewritten as
hotels = pd.DataFrame([2, 3, 4], index=['a','b','c'], columns=['v'])
# can't start and end with same hotel

nights = hotels.values.sum() # 9
available_hotels = hotels.index # [0,1,2]
num_available_hotels = len(available_hotels) # 3
nights_per_hotel = hotels.values.max() # 4

orig_path = []
_max = hotels[hotels['v'] == hotels.values.max()]
_min= hotels[hotels['v'] == hotels.values.min()]
min_max = _min.append(_max)
min_max = min_max.sort_values(by=['v'], ascending=True)

hotels = hotels[~hotels.index.isin(min_max.index)]
remainder = abs(int(min_max.diff().dropna()['v']))
while min_max.values.sum() != remainder:
    for hotel in min_max.iterrows():
        if hotel[1]['v'] == 0:
            pass
        else:
            hotel[1]['v'] -= 1
            orig_path.append(hotel[0])

min_max = min_max[min_max!=0].dropna()
hotels = hotels.append(min_max)
hotels = hotels.sort_values(by=['v'], ascending=False)
remainder = 0
while hotels.values.sum() != remainder:
    for hotel in hotels.iterrows():
        if hotel[1]['v'] == 0:
            pass
        else:
            hotel[1]['v'] -= 1
            orig_path.append(hotel[0])

            
prev = 'z'
exception = []
for p in orig_path:
    if p == prev:
        exception.append(p)
    prev = p
    

# so for [10,15,20]
# repeat orig_path 3 times
prev = 3
exception = []
_3_path = orig_path * 3
for p in _3_path:
    if p == prev:
        exception.append(p)
    prev = p
    

#if len(exception) > 0:
    #print('not possible used same hotel at least twice in a row')
    #print(exception)
    #print(_3_path)
#else:
    #print("possible")
    #print(_3_path)
```

```python
# Books % 3 = x groups 2 remain
# Books % 4 = x groups 3 remain
# Books % 5 = x groups 4 remain

N = []
for book in range(100):
    n_divisible_by_book = []
    if book % 3 == 2:
        n_divisible_by_book.append(3)
    if book % 4 == 3:
        n_divisible_by_book.append(4)
    if book % 5 == 4:
        n_divisible_by_book.append(5)
    if len(n_divisible_by_book) == len([3,4,5]):
        N.append(book)


print(N[0],"books")
```
